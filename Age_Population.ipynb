```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import files
```

```python
uploaded = files.upload()

# Load the Dataset
file_name = 'train.csv'  # Update if the uploaded file name is different
titanic_data = pd.read_csv(file_name)
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=titanic_data, x='Sex', palette='pastel')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Histogram for age distribution
plt.figure(figsize=(10, 6))
sns.histplot(titanic_data['Age'].dropna(), bins=30, kde=True, color='skyblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bar chart for passenger class distribution
plt.figure(figsize=(10, 6))
sns.countplot(data=titanic_data, x='Pclass', palette='muted')
plt.title('Passenger Class Distribution')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()
```

```python
# Bar chart for survivors distribution
plt.figure(figsize=(10, 6))
sns.countplot(data=titanic_data, x='Survived', palette='Set2')
plt.title('Survivor Distribution')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
# Bar chart for embarkation point distribution
plt.figure(figsize=(10, 6))
sns.countplot(data=titanic_data, x='Embarked', palette='deep')
plt.title('Embarkation Point Distribution')
plt.xlabel('Embarkation Point')
plt.ylabel('Count')
plt.show()
```
